```python
from dataclasses import dataclass, field
from datetime import datetime
from typing import List, Tuple
from collections import defaultdict

# Helper function to convert time string to seconds
def time_to_seconds(time_str: str) -> int:
    t = datetime.strptime(time_str, "%M:%S")
    return t.minute * 60 + t.second

@dataclass
class ChapterTranscript:
    chapter: str
    timestamp: str
    transcript: List[str] = field(default_factory=list)

    @staticmethod
    def parse_chapters(chapters: List[str]) -> List[Tuple[int, 'ChapterTranscript']]:
        """Parse chapters into ChapterTranscript objects with timestamps in seconds."""
        return [
            (time_to_seconds(chapter.split(' - ')[0]), 
             ChapterTranscript(
                 chapter=chapter.split(' - ')[1], 
                 timestamp=chapter.split(' - ')[0]
             ))
            for chapter in chapters
        ]

    @staticmethod
    def assign_transcripts_to_chapters(
        chapter_data: List[Tuple[int, 'ChapterTranscript']],
        transcript_data: List[Tuple[float, str]]
    ) -> List['ChapterTranscript']:
        """Assign transcript entries to the appropriate chapters."""
        current_chapter_index = 0
        for transcript_time, transcript_text in transcript_data:
            # Move to the next chapter if the transcript time exceeds the current chapter time
            while (current_chapter_index + 1 < len(chapter_data) and 
                   transcript_time >= chapter_data[current_chapter_index + 1][0]):
                current_chapter_index += 1

            # Add the transcript to the current chapter
            chapter_data[current_chapter_index][1].transcript.append(transcript_text)

        # Return only the ChapterTranscript objects
        return [chapter for _, chapter in chapter_data]


chapter_data = ChapterTranscript.parse_chapters(chapters)
transcript_data = [(float(t.split(' - ')[0]), t.split(' - ')[1]) for t in cleaned]

# Assign transcripts to chapters
merged_chapters = ChapterTranscript.assign_transcripts_to_chapters(chapter_data, transcript_data)

# Display results
for chapter in merged_chapters:
    print(f"Chapter: {chapter.chapter} (Timestamp: {chapter.timestamp})")
    print("Transcripts:")
    for line in chapter.transcript:
        print(f"  - {line}")
    print()
```
